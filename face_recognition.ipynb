```python
import cv2

# Load image (in original format)
image = cv2.imread('./Dataset/Due Lipa/464400_1.3.jpg', cv2.IMREAD_UNCHANGED)  # Use IMREAD_COLOR for color images

# Check if loaded
if image is None:
    print("Error: Could not load image.")
else:
    # Display the image "as it is" in an OpenCV window
    cv2.imshow('Original Image', image)
    cv2.waitKey(0)  # Wait for a key press
    cv2.destroyAllWindows()
```

```python
plt.imshow(image_rgb)
plt.title('Color Image (RGB)')
plt.axis('off')
plt.show()
```
